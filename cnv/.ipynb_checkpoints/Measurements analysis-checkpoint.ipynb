# 3. Correlation Matrix

```python
corr_features = measurements.copy()
corr_features.drop(['Puzzle','Strategy'], axis=1)
f,ax = plt.subplots(figsize=(20, 20))
sns.heatmap(corr_features.corr(), annot=True, linewidths=.9, fmt= '.2f',ax=ax)
```

## Notes

__1):__ Number of __instructions__ is __highly correlated with number of blocks (0.99)__. Thus, __we will not comment on number of blocks__. Also, there is a __correlation of 0.99 between instructions and time (ms)__, so we will assume that __time of execution is determined by the number of instructions__.

__2):__ __Area__ is __highly correlated__ with __regular load (0.78)__ and __regular store (0.87)__, which indicates that __as these values increase, bigger chunks of memory are being allocated__. 

__3): Missing elements__ __highly correlated__ with __instructions (0.89)__. Thefore, the more incomplete the puzzles are, the more __computationally demanding is to solve it__. Also, it is __highly correlated__ with __regular store__ (TODO: WHY??).

# 5. Mean values grouped on missing elements

```python
measurements.groupby(['Missing ele','Strategy'], as_index=True)['Instructions','Methods',
                                                 'Field load','Field store', 'Regular load','Regular store',
                                                'new', 'newarray','Time (ms)'].mean()
```

## Notes:

Recall that (9,16,22) belong to the 9x9 puzzle, (31,54) to the 16x6 and (64,114) to the 25x25. 

__1)__ __Increasing missing elements__ has a __high impact in # instructions for the CP and BFS__ algorithm (__doubling missing elements__ almost __doubles instructions__).

__2)__ __Increasing missing elements__ leads to an __increase of field load, regular load and regular store__. However, this __does not happen for the DLX solver__.

__3) Increasing missing elements__ leads to an __increase of methods for the CP and BFS__, specially in the __16x16__ puzzle.
